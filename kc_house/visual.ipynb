```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import regression as reg
import matplotlib.pyplot as plt
```

```python
fig = plt.gcf()
plt.figure(figsize=(8, 6), dpi=100)

def plot_multiple(col, x_label, y_label, title, name):
    gd = pd.read_csv('../data/output/lf_n_grn_gd_2e-12.csv')
    newton = pd.read_csv('../data/output/lf_n_grn_newton_0.99.csv')
    plt.plot(range(10), gd[col][0:10], label='gd')
    plt.plot(range(10), newton[col][0:10], label='newton')
    plt.xlim(0, 10)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    plt.show()
    plt.savefig(name, dpi=1000)
```

```python
plot_multiple('loss_func_list', 'count', 'loss function', 'loss function vs count', '../data/output/b1.png')
```
