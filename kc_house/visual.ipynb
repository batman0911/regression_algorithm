```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import regression as reg
import matplotlib.pyplot as plt
```

```python
# fig = plt.gcf()
# plt.figure(figsize=(30, 6), dpi=100)

def plot_multiple(col, x_label, y_label, title, name, min, max):
    plt.figure(figsize=(4, 6), dpi=100)
    gd = pd.read_csv('../data/output/lf_n_grn_gd_1.csv')
    bgd = pd.read_csv('../data/output/lf_n_grn_bgd_128.csv')
    agd = pd.read_csv('../data/output/lf_n_grn_agd_0.5.csv')
    agd_bt = pd.read_csv('../data/output/lf_n_grn_agd_bt_3.csv')
    newton = pd.read_csv('../data/output/lf_n_grn_newton_1.csv')
    plt.plot(range(min, max), gd[col][min:max], label='gd')
    plt.plot(range(min, max), bgd[col][min:max], label='bgd')
    plt.plot(range(min, max), agd[col][min:max], label='agd')
    plt.plot(range(min, max), agd_bt[col][min:max], label='agd_bt')
    plt.plot(range(min, max), newton[col][min:max], label='newton')
    plt.xlim(min-1, max)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    # plt.show()
    plt.savefig(name, dpi=800)
```

```python
min = 300
max = 400
plot_multiple('loss_func_list', 'count', 'loss function', 'loss function vs count', '../data/output/b_' + str(min) + '_' + str(max), min, max)
```

```python
min = 300
max = 400
plot_multiple('grad_norm_list', 'count', 'gradient norm', 'gradient norm vs count', '../data/output/g_' + str(min) + '_' + str(max), min, max)
```
