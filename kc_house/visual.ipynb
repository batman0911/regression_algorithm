```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import regression as reg
import matplotlib.pyplot as plt
```

```python
# fig = plt.gcf()
# plt.figure(figsize=(30, 6), dpi=100)

def plot_multiple(col, x_label, y_label, title, name):
    plt.figure(figsize=(6, 3), dpi=100)
    gd = pd.read_csv('../data/output/lf_n_grn_gd_2e-12.csv')
    agd = pd.read_csv('../data/output/lf_n_grn_agd_3e-12.csv')
    agd_bt = pd.read_csv('../data/output/lf_n_grn_agd_bt_5.129727980068933e-13.csv')
    newton = pd.read_csv('../data/output/lf_n_grn_newton_0.99.csv')
    plt.plot(range(10), gd[col][0:10], label='gd')
    plt.plot(range(10), agd[col][0:10], label='agd')
    plt.plot(range(10), agd_bt[col][0:10], label='agd_bt')
    plt.plot(range(10), newton[col][0:10], label='newton')
    plt.xlim(-1, 10)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.legend()
    # plt.show()
    plt.savefig(name, dpi=800)
```

```python
plot_multiple('loss_func_list', 'count', 'loss function', 'loss function vs count', '../data/output/b1.png')
```
